# Build U-Net example

this example shows how to build a u-net model from the code given

it depends on tensorflow and keras package.

it was tested on tensorflow 1.13 and 1.14 only! and may not work in tensorflow 2.0 +

I used config. information (i.e., json file) to control the network structure and the training process, so that I can run many tasks one after the other automatically.

Becasue of this, the config. information shown in this file contains many redundant information that has nothing to do with building the CNN network itself, I keep those information just in case you need them to understand other part of the codes (which are very messy, and I am sorry about that)

```python
import unet
```

Here is an example of defining the u-net via the config information

### note that this information controls the encoder part only, the decoder part is automatically generated by mirroring the given information

```python
config_1={'type':'parent_folder_name',
        'name':'model_name',
        'net_filter':[32,32,32,64,64], # filter size (i.e., num. of image channels) of each layer (conv. or max-pool)
        'net_kernel':[[3,3],[3,3],[2,2],[3,3],[3,3]], # kernel (row and col.) size of each layer (conv. or max-pool)
        'net_stride':[[1,1],[1,1],[2,2],[1,1],[1,1]], # stride (row and col.) size of each layer (conv. or max-pool)
        'net_pool_pos': 3, # an integer, here it means every 3 layer is a max-pooling layer, remove this entry to build network without any pooling layer (line 385 in unet.py)
        'net_u_connections':True, # add skip connections between encoder and decoder (conv. or max-pool)
        'net_height_in_prediction':False, # predict terrain elevation + water depth, when set to positive float (denoted as val), the prediction will be (wd+ele)*val (not working, tested already), this has nothing to do with building the CNN
        'data_root':'folder_path', # a folder path that indicates the file location of dataset, this has nothing to do with building the CNN
        'data_resolution':'give string here', # data's spatial resolution (this info. is used to generate filename), this has nothing to do with building the CNN
        'data_features':'give string here', # data's features (i.e., how many features were pre-computed, this info. is used to generate filename), this has nothing to do with building the CNN
        'data_patch_size':256, # patch size sampled from the terrain elevation data (i.e., CNN's input size)
        'data_random_seed':1, # random seed that shuffles the dataset, this has nothing to do with building the CNN
        'data_propotion':3, # ratio between training and test dataset, this has nothing to do with building the CNN
        'data_output_channels':2, # outpu channel of the CNN
        'data_patch_ratio':2, # how many patches to sample for each catchment area, this has nothing to do with building the CNN
        'data_augmentation':True, # data augmentation for training
        'data_aug_rot':60, # False: no rotation, integer: with rotation (degrees), this has nothing to do with building the CNN
        'data_aug_flip_x':[-1, 1], # flip data along x directions, this has nothing to do with building the CNN
        'data_aug_flip_y':[-1, 1], # flip data along y directions, this has nothing to do with building the CNN
        'data_feature_extractor':0, # integer that indicates which feature extractor is used, this has nothing to do with building the CNN
        'train_batch_size':2, # training batch size, this has nothing to do with building the CNN
        'train_epoch':200, # training epoch, this has nothing to do with building the CNN
        'train_learning_rate':0.00005, # learning rate, this has nothing to do with building the CNN
        'train_optimizer':'adam', # gradient descent optimizer
        'train_weighted_loss_s':False, # set a number to activate weighted loss, this changes the loss function
        'train_weighted_loss_c':False # set a number to activate weighted loss, this changes the loss function
       }
```

```python
model_1=unet.build(config_1, 2) # build u-net model using config_1 with input channel = 2
```

```python
model_1.summary()
```

here is an example in which I introduced new keyword for controlling the u-net structure.

u-net consists of many components, each of conponent is a sequence of n conv. layer and 1 max-pooling layer, i.e., [conv. conv. ... conv. max.]

net_stack_repeat is a list of two items, net_stack_repeat[0] defines the n mentioned above, net_stack_repeat[1] defines the num. of conv. layer at the end of the encoder part.<br>

e.g., net_stack_repeat = [3,2] gives you: [conv. conv. conv. max.] , [..], ...., [conv. conv. conv. max.], [conv. conv.]

if net_stack_repeat is given, then <b>net_filter should be given as a list of integers</b> which defines the filter size of each components, as well as how many components will be created.

e.g., net_stack_repeat=[2,1], net_filter=[8,16] gives you [conv. conv. max.]\(filter=8\), [conv.]\(filter=16\)

if net_stack_repeat is given, <b>net_kernel would be an integer</b> that controls all conv. layer (i.e., all layer has the same kernal size)

if net_stack_repeat is given, <b>net_stride would be an integer</b> that controls all max-pooling layer (i.e., all layer has the same kernal size and stride size)

see check_net_parameters() in unet.py for details

```python
config_2={'type':'parent_folder_name',
        'name':'model_name',
        'net_stack_repeat':[2,1], # (see line 321 in unet.py)
        'net_filter':[8,16,32,64,128,256,512], # see above
        'net_kernel':7, # see above
        'net_stride':2, # see above
        'net_u_connections':True, # add skip connections between encoder and decoder
        'net_height_in_prediction':False, # predict terrain elevation + water depth, when set to positive float (denoted as val), the prediction will be (wd+ele)*val (not working, tested already), this has nothing to do with building the CNN
        'data_root':'folder_path', # a folder path that indicates the file location of dataset, this has nothing to do with building the CNN
        'data_resolution':'give string here', # data's spatial resolution (this info. is used to generate filename), this has nothing to do with building the CNN
        'data_features':'give string here', # data's features (i.e., how many features were pre-computed, this info. is used to generate filename), this has nothing to do with building the CNN
        'data_patch_size':1024, # patch size sampled from the terrain elevation data (i.e., CNN's input size)
        'data_random_seed':1, # random seed that shuffles the dataset, this has nothing to do with building the CNN
        'data_propotion':3, # ratio between training and test dataset, this has nothing to do with building the CNN
        'data_output_channels':2, # outpu channel of the CNN
        'data_patch_ratio':2, # how many patches to sample for each catchment area, this has nothing to do with building the CNN
        'data_augmentation':True, # data augmentation for training
        'data_aug_rot':60, # False: no rotation, integer: with rotation (degrees), this has nothing to do with building the CNN
        'data_aug_flip_x':[-1, 1], # flip data along x directions, this has nothing to do with building the CNN
        'data_aug_flip_y':[-1, 1], # flip data along y directions, this has nothing to do with building the CNN
        'data_feature_extractor':0, # integer that indicates which feature extractor is used, this has nothing to do with building the CNN
        'train_batch_size':2, # training batch size, this has nothing to do with building the CNN
        'train_epoch':200, # training epoch, this has nothing to do with building the CNN
        'train_learning_rate':0.00005, # learning rate, this has nothing to do with building the CNN
        'train_optimizer':'adam', # gradient descent optimizer
        'train_weighted_loss_s':False, # set a number to activate weighted loss, this changes the loss function
        'train_weighted_loss_c':False # set a number to activate weighted loss, this changes the loss function
       }
```
